```python
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np

from tri_ct_tools.image.checker import plot_intensity
from tri_ct_tools.image.reader import singlecam_series

%matplotlib widget
```

```python
def mean_prog_plot(
        img_folder: Path,
        frames=range(50, 1350),
        points=[
            [200, 600],
            [250, 600],
            [200, 550],
            [250, 550]
            ]
        ):
    images = singlecam_series(img_folder, frames, (450, 1548))
# print(images.shape)
# First point: mean from 0 to 1 = images[point, 0:1].mean()
# second point: mean from 0 to 2 = images[point, 0:1].mean()
#               : from 0 to 3
    fig, axs = plt.subplots(2, 1, layout='constrained')
    means = []
    for point in points:
        mean = [images[0:i, *point].mean() for i in range(2, len(frames))]
        rel_mean = mean / mean[-1]
        axs[0].plot(frames[2:], np.array(mean).astype(np.int16), label=f"y: {point[0]}, x: {point[1]}")
        axs[1].plot(frames[2:], rel_mean, label=f"y: {point[0]}, x: {point[1]}")

    axs[0].set_title("Progression of mean pixel value")
    axs[0].set_ylabel("Counts")
    axs[0].set_xlabel("Frames")

    axs[1].set_title("Progression of relative mean")
    axs[1].set_ylabel("Mean value over final mean")
    axs[1].set_xlabel("Frames")

    fig.suptitle(img_folder.name)
    plt.legend()
```

```python
img_folder= Path(R"s:\NH4Cl_0gl_150lmin_Scatter_S12D3\camera 3")

frames = range(50, 1350)
points = [
    [200, 600],
    [250, 600],
    [200, 550],
    [250, 550]
]

mean_prog_plot(img_folder, frames, points)
```
